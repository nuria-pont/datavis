```python
any=[]
index=[]
x=0
for i in (df_pax_final['Dat']):
    x=x+1
    any.append(int(i[:4]))
    index.append(x)
df_any = pd.DataFrame (any,columns=['Any'])
df_pax_final=pd.concat([ df_any,df_pax_final], axis=1)
df_pax_final.head(3)
```

```python
#preparem descripcions
```

```python
df_GCh_Al = pd.DataFrame (columns=['GCh_Al'])
df_pax_final=pd.concat([ df_GCh_Al,df_pax_final], axis=1)
df_pax_final.head(3)
```

```python
# alies GCh --> 0:nul.la menció; 1: retòrica ; 2: contra la discriminació; 3: substantives/substancial . 

for i in range(len(df_pax_final)):
    val=df_pax_final.iloc[i]["GCh"]
    if val==1:
        r = "Retòrica"
    elif val==2:
        r = "Contra la discriminació"
    elif val==3:
        r = "Substantives/substancial";
    else:
        r = "----"    
    df_pax_final["GCh_Al"][i]=r
```
